```python
%matplotlib inline
plt.xlabel("lot size")
plt.ylabel("price")
plt.scatter(df.lotsize,df.price, color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["lotsize"]],df.price) #fitting the data in linear regrssion model assining the x axis and y axis 
```

```python
reg.predict(np.array([[1500]]))  # Adding an extra pair of brackets makes it 2D
```

```python
reg.coef_ #m
```

```python
reg.intercept_ #c
```
